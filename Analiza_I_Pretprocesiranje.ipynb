Cilja promenljiva je svakako Total.Cup.Points koja uzima vrednosti iz intervala [0,100] i predtsavlja sumu ocena iz razlicitih kategorija, kao sto su: sweetness, balance, acidity, body i sl. 
Cilj nam je da predvidimo ocenu kafe na osnovu drugih atributa (nadmorska visina, vrsta kafe, defekti, poreklo...) tako da cemo izbaciti nepotrebne kolone (pored onih koje ucestvuju u odredjivanju vrednosti kolone Total.Cup.Points, imamo i druge nepotrebne kolone - Certification.Contact, Lot.Number ... )

```python
coffee.columns
```

```python
coffee_subset=coffee[['Species','Owner','Country.of.Origin','Farm.Name','Mill','Company','Region','Harvest.Year','Variety','Processing.Method','Category.One.Defects','Quakers','Color','Category.Two.Defects','altitude_mean_meters','Total.Cup.Points']]
```

```python
coffee_subset
```

```python
#Vec vidimo da u skupu postoje nedostajuce vrednosti. Izborjimo ih
coffee_subset.isna().sum()
```

```python
na=coffee_subset.isna().sum()

plt.figure(figsize=(10,8))
plt.bar(x=na.index,height=na.values)
plt.grid(visible=True,alpha=0.1)
plt.title("Nedostajuce vrednosti")
plt.xticks(rotation=90)
plt.yticks(np.arange(0,380,30))
plt.show()
```

```python
# Raspodela numerickih atributa
_=pd.DataFrame.hist(coffee_subset,figsize=(10,10),column=coffee_subset.columns,bins=10,xlabelsize=10,ylabelsize=10)
```

```python
# kategirski atributi 
categorical = coffee_subset.select_dtypes(include=['object','category']).columns
categorical
```

```python
#zameljujemo sve kategoricke vrednosti u numercike (osim NaN)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for c in categorical:
    idx=~coffee_subset[c].isna()
    coffee_subset.loc[idx,c]=le.fit_transform(coffee_subset.loc[idx,c])

coffee_subset.head()
```

```python
import math
plt.figure(figsize=(10,10),layout='tight')
idx=1
for c in categorical:
   plt.subplot(5,math.ceil(categorical.size/5),idx)
   idx+=1
   plt.bar(coffee_subset[c].dropna().unique(),coffee_subset[c].value_counts())
   plt.grid(visible=True, alpha=0.6)
   plt.title(c)

plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(30,25))
sns.heatmap(data=coffee_subset.corr(),annot=True)
```

Pre nego sto popunimo nedostajuce vrednosti mozemo proveriti korelaciju
Na osnovu heat mape matrice korelacije najveci koeficijent korelacije ima par [Owner,Company] sto nije iznenadjujuce, mada i ta najveca vrednost od 0.46 nije dovoljno dobra da bi na osnovu korelacija atributa popunjavali nedostajuce vrednosti 

```python
#Popunimo nedostajuce vrednosti
# U numerickim atributima Quakers i altitude_mean_meters imaju nedostajuce vrednosti - popunicemo ih prosecnom vrednoscu
coffee_subset['Quakers'].fillna(value=round(coffee_subset['Quakers'].mean()),inplace=True)
coffee_subset['altitude_mean_meters'].fillna(value=coffee_subset['altitude_mean_meters'].mean(),inplace=True)
```

```python
# Nedostajuce vrednosti kategorickih atributa cemo popunjavati sa vrednostima koje se najcesce pojavljuju
missing_categorical= [var for var in categorical if coffee_subset[var].isna().sum()>0]
for var in missing_categorical:
    coffee_subset[var]=coffee_subset[var].fillna(coffee_subset[var].mode()[0])
```
