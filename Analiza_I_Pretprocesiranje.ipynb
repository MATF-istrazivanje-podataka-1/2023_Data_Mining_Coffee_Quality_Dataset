```python
coffee_subset=coffee[['Species','Owner','Country.of.Origin','Farm.Name','Mill','Company','Region','Harvest.Year','Variety','Processing.Method','Category.One.Defects','Quakers','Color','Category.Two.Defects','altitude_mean_meters','Total.Cup.Points']]
```

```python
# kategirski atributi 
categorical = coffee_subset.select_dtypes(include=['object','category']).columns
categorical
```

```python
# Pogledajmo podatke u kategoriskim atributima - da li im je potrebno sredjivanje za lakse razumevanje 
for i in categorical:
    col=coffee_subset[i]
    unique=col.unique()
    
    print(unique)
```

Primetimo da kolone poput Mill, Farm.Name, Company i Owner imaju previse razlicitih vrednosti koje nece pomoci u klasifikovanju kvaliteta kafe - mogu dovesti do preprilagodjavanja na ovako malom skupu podataka
Dodatno je interesantna kolona Harvest.Year - vrednosti su formatirane na razlicite nacine, s tim sto kod nekih i fali informacija o godini 
-ako fali informacija o godini postavicemo vrednost na NaN i posle je popuniti
- ako je datum formatiran kao god1 / god2, ostavimo samo god2 
- i sl. da bi u svakom polju pisala samo jedna godina, a ne preciziniji datum 


```python

coffee_subset["Harvest.Year"].replace(to_replace=r'^[a-zA-z -]*$',value=np.nan,inplace=True,regex=True)
coffee_subset["Harvest.Year"].replace(to_replace=r'.*(\d{4})$',value='\\1',inplace=True,regex=True)
coffee_subset["Harvest.Year"].replace(to_replace=r'.*/(\d{4})',value='\\1',inplace=True,regex=True)
coffee_subset["Harvest.Year"].replace(to_replace=r'.*/(\d{2}).*',value="20"+'\\1',inplace=True,regex=True)
coffee_subset["Harvest.Year"].replace(to_replace=r'.*\b(\d{4})\b.*',value='\\1',inplace=True,regex=True)
```

```python
#zameljujemo sve kategoricke vrednosti u numercike (osim NaN)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for c in categorical:
    idx=~coffee_subset[c].isna()
    coffee_subset.loc[idx,c]=le.fit_transform(coffee_subset.loc[idx,c])

coffee_subset.head()
```

```python
import math
plt.figure(figsize=(10,10),layout='tight')
idx=1
for c in categorical:
   plt.subplot(5,math.ceil(categorical.size/5),idx)
   idx+=1
   plt.bar(coffee_subset[c].dropna().unique(),coffee_subset[c].value_counts())
   plt.grid(visible=True, alpha=0.6)
   plt.title(c)

plt.show()
```

Nemamo neke izrazene korelacije, i kao sto je predpostavljeno te vrednosti su veoma bliske nuli za atribute Mill, Company, Farm.Name i Owner i njih cemo izbaciti iz skupa

```python
nepotrebne=["Mill","Company","Farm.Name","Owner"]
for col in nepotrebne:
    coffee_subset=coffee_subset.drop(col, axis=1)

coffee_subset.head()
```

```python
#Popunimo nedostajuce vrednosti
# U numerickim atributima Quakers i altitude_mean_meters imaju nedostajuce vrednosti - popunicemo ih prosecnom vrednoscu
coffee_subset['Quakers'].fillna(value=round(coffee_subset['Quakers'].mean()),inplace=True)
coffee_subset['altitude_mean_meters'].fillna(value=coffee_subset['altitude_mean_meters'].mean(),inplace=True)
```

```python
# Nedostajuce vrednosti kategorickih atributa cemo popunjavati sa vrednostima koje se najcesce pojavljuju
categorical = coffee_subset.select_dtypes(include=['object','category']).columns

missing_categorical= [var for var in categorical if coffee_subset[var].isna().sum()>0]
for var in missing_categorical:
    coffee_subset[var]=coffee_subset[var].fillna(coffee_subset[var].mode()[0])
```

Stavila sam 30 korpica na histogram da bi dosla do izrazaja 'normalna' rasporedjenost ocena - nije zaista, bar ne ako gledamo ceo interval ocena [0,100]
Probajmo da diskretizujemo po istoj sirini intervala

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
kbd= KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform', subsample=None)
# diskretizujemo po sirini intevrala


Points=kbd.fit_transform(pd.DataFrame(coffee_subset['Total.Cup.Points']))
```

```python
coffee_subset['Total.Cup.Points']=Points
```
